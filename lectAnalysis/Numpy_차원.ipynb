# 0-차원 배열 (스칼라, Scalar)

```python
import numpy as np
arr = np.array(12)

print(arr)              # 출력 : 12
print(np.shape(arr))    # 출력 : ( ) → 0-차원
```

# 1-차원 배열 (벡터, Vector)

```python
import numpy as np
arr = np.array([2, 4, 6, 8, 10])    # 1차원 배열

print(arr)              # 출력 : [ 2  4  6  8 10]
print(np.shape(arr))    # 출력 : 5,
```

```python
import numpy as np

print(np.zeros(10))                # 출력 : [0. 0. 0. 0. 0.] → 0 으로 채워진 1차원 배열 생성
print(np.ones(7))                 # 출력 : [1. 1. 1. 1. 1.] → 1 로 채원진 1차원 배열 생성
print(np.empty(5))                # 출력 : [1. 1. 1. 1. 1.] → 빈 배열 생성, 메모리 상태에 따라 값이 달라진다.
print(np.full(9, 6))              # 출력 : [7 7 7 7 7] →7 로 5 개 채워진 1차원 배열 생성
print(np.arange(5))               # 출력 : [0 1 2 3 4] → 0 부터 시작하는 5 개의 숫자 생성 (0, 1, 2, 3, 4)
print(np.arange(2,20,3))          # 출력 : [ 2  5  8 11 14 17] → 일정한 간격의 배열 생성, (시작, 정지, 단계)
print(np.random.random(3))        # 출력 : 0 이상 1 미만의 난수 3개를 랜덤하게 배열 생성
print(np.linspace(0,100, num=10))  # 출력 : [  0.  50. 100.] → 선형 간격으로 배치된 값으로 배열 생성
```

1-차원 배열을 한꺼번에 계산하는 벡터화 연산(vectorized operation)

```python
import numpy as np
arr1 = np.array([1, 3, 5])      # 1차원 배열
arr2 = np.array([2, 4, 6])      # 1차원 배열

print(arr1 * 2)       # 출력 : [2, 6, 10]
print(arr1 + arr2)    # 출력 : [3, 7, 11]
```

# 2-차원 배열 (행렬, Matrix)

```python
import numpy as np
arr = np.array([[2, 4, 6], [1, 3, 5]])    # 2행 3열, 2차원 배열

print(arr)
print(np.shape(arr))    # 출력 : (2, 3) → 2행 3열
```

```python
import numpy as np
arr = np.array([[2, 4], [6, 8], [10, 12]])

print(arr)              # 출력 : 3행 2열 데이터
print(np.shape(arr))    # 출력 : (3, 2) → 3행 2열

print(len(arr))         # 출력 : 3 → 행의 개수
print(len(arr[0]))      # 출력 : 2 → 열의 개수
```

```python
import numpy as np
np.arange(6).reshape(2, 3)
```

넘파이에서는 2차원 배열을 생성하면서 초기값을 지정할 수 있다.
이를 '배열을 초기화한다'고 하며 np.zeros((행, 열)) 처럼 배열 전체를 0으로 채우는 것이 대표적

```python
import numpy as np

print(np.zeros((2, 3)))             # 출력 : 2행 3열의 0으로 채워진 배열 생성
print(np.ones((2, 3)))              # 출력 : 2행 3열의 1로 채워진 배열 생성
print(np.empty((2, 3)))             # 출력 : 2행 3열의 "초기화되지 않은" 배열 (값은 메모리에 따라 달라짐)
print(np.full((2, 3), 7))           # 출력 : 2행 3열이 모두 7로 채워진 배열 생성
print(np.arange(6).reshape(2, 3))   # 출력 : 0부터 시작하는 6개의 정수를 2행 3열로 재배열
print(np.arange(2, 20, 3).reshape(2, 3))         # 출력 : 2부터 시작해서 3씩 증가하는 수를 2행 3열 배열로 구성
print(np.random.random((2, 3)))     # 출력 : 2행 3열의 난수를 0 이상 1 미만 범위로 랜덤 생성
print(np.linspace(0, 100, num=6).reshape(2, 3))  # 출력 : 0부터 100까지 균등하게 나눈 6개 숫자를 2행 3열로 재배열
```

# 3-차원 배열 (텐서, Tensor)

3-차원 배열을 텐서(Tensor)라고 하며, 2-차원 배열을 여러 개 겹쳐진 구조.
3-차원 배열은 깊이(depth), 행, 열로 구성되며 인공지능에서 자주 사용되는 텐서(Tensor)의 기본 구조.
아래와 같이 2차원 배열을 2번 사용하면 해당 3-차원 배열의 깊이는 2가 된다.
[ [ 행, 행, 행 ] [ 열, 열, 열 ] ], [ [ 행, 행, 행 ] [ 열, 열, 열 ] ]
└----------- 깊이 -----------┘ └----------- 깊이 -----------┘

```python
import numpy as np
arr = np.array([[[111, 112, 113], [121, 122, 123]],
                [[211, 212, 213], [221, 222, 223]]])    # 3차원 배열

print(arr)              # 출력 : 3차원 데이터
print(np.shape(arr))    # (2, 2, 3) → 깊이 2, 행 2, 열 3
```

```python
import numpy as np
arr = np.array([[[111, 112, 113], [121, 122, 123]],
                [[211, 212, 213], [221, 222, 223]]])    # 3차원 배열

print(len(arr))         # 출력 : 2 →깊이
print(len(arr[0]))      # 출력 : 2 →행의 개수
print(len(arr[0][0]))   # 출력 : 3 →열의 개수
```
