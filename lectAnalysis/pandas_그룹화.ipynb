# 데이터 그룹화

```python

import pandas as pd
df =  pd.read_csv('./pandas_groupby.csv', encoding='CP949')
df.head(2)
```

```python
df['지역'].unique()     # '지역' 컬럼에는 4개의 지역이 저장되어 있음을 알 수 있다.
```

```python
df.groupby(by='지역')     # 'by=' 매개변수는 생략 가능하다.
```

```python
df.groupby('지역').count()      # mean() 함수는 에러 발생, 평균은 숫자형 컬럼에서 사용한다.
```

```python
df.groupby('지역')['판매개수'].mean()      # 데이터프레임 형태로 '지역'별 판매개수의 평균을 표시한다.
```

다중 기준으로 그룹화

```python
df.groupby(['지역', '제품'])['판매개수'].sum()
```

```python
df.groupby(['판매원', '제품'])[['판매개수','판매액']].sum()
```

```python
df.groupby(['판매원', '제품']).agg({'판매개수': 'sum','판매액': 'mean'})
```

```python
result = df.groupby(['판매원', '제품'])[['판매액']].sum()
result = result.rename(columns={'판매액':'합계'})
result
```

```python
df[['년', '월', '일']] = df['날짜'].str.split('-', expand=True)
df.groupby(['년', '월', '판매원']).agg({'판매개수': 'sum', '판매액': 'mean'}) #
# df
```
