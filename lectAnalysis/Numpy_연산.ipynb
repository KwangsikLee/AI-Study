# add() 메서드

```python
import numpy as np
a = [21, 22, 23, 24, 25, 87]
b = [11, 12, 13, 14, 15, 0]

dap = np.add(a, b)
dap     # 출력 : [32, 34, 36, 38, 40]
```

# subtract() 메서드

```python
import numpy as np
a = [21, 22, 23, 24, 25]
b = [11, 12, 13, 14, 15]

dap = np.subtract(a, b)
dap   # 출력 : [10, 10, 10, 10, 10]
```

# multiply() 메서드

```python
import numpy as np
a = [21, 22, 23, 24, 25]
b = [11, 12, 13, 14, 15]

dap = np.multiply(a, b)
dap     # 출력 : [231, 264, 299, 336, 375]
```

# divide() 메서드

소수점 자리수 설정 : result = np.round(dap, 2)

```python
import numpy as np
a = [20, 21, 24, 30, 35]
b = [4, 3, 3, 5, 7]

dap = np.divide(a, b)
dap     # 출력 : [5, 7, 8, 6, 5]
```

# mod(), remainder() 메서드

mod(), remainder() 메서드

```python
import numpy as np
a = [10, 20, 30, 40, 50]
b = [1, 2, 3, 4, 5]

dap = np.mod(a, b)
dap     # 출력 : [0, 0, 0, 0, 0]
```

```python
import numpy as np
a = [11, 21, 31, 41, 51]
b = [1, 2, 3, 4, 5]

dap = np.remainder(a, b)
dap     # 출력 : [0, 1, 1, 1, 1]
```

# divmod() 메서드

divmod() 메서드는 나눗셈의 몫과 나머지를 모두 반환한다.
반환 값은 두 개의 배열이며, 첫 번째 배열에는 몫이,
두 번째 배열에는 나머지가 저장된다.
dap = np.divmod(a, b)
dap[1]　　　# 나머지만 출력

```python
import numpy as np
a = [11, 21, 31, 41, 51]
b = [1, 2, 3, 4, 5]

dap = np.divmod(a, b)
dap     # 출력 : array([11, 10, 10, 10, 10], array([0, 1, 1, 1, 1])
```

# power() 메서드

power() 메서드는 첫 번째 배열의 값을 두 번째 배열 값으로 거듭제곱

```python
import numpy as np
a = [1, 2, 3, 4, 5]
b = [1, 2, 3, 4, 5]

dap = np.power(a, b)
dap     # 출력 : [1, 4, 27, 256, 3125]
```

# abs(), absolute() 메서드

abs()와 absolute()메서드 모두 배열의 값을 절대값

```python
import numpy as np
a = [1, -2, 3, -4, 5]

dap = np.abs(a)
dap     # 출력 : [1, 2, 3, 4, 5]
```

# 적용

```python
import numpy as np
a = np.array([1, 2, 3, 4, 5])
b = np.ones(5)        # [1. 1. 1. 1. 1.] → 실수형 배열

print(np.add(a, b))   # 출력 : [2. 3. 4. 5. 6.]
```

```python
import numpy as np
a = np.array([[1, 2, 3],
              [4, 5, 6]])
b = np.full((2, 3), 10)     # shape = (2행, 3열) → 10으로 채움

dap = np.add(a, b)
print(dap)
```

```python
import numpy as np
a = np.array([1, 2, 3, 4, 5])

print(np.multiply(a, 2))   # 출력 : [ 2  4  6  8 10]
```
