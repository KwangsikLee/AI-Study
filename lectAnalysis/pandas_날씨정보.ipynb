```python
import pandas as pd
import numpy as np
```

```python
# ì§€ì—­ê³¼ ë‚ ì§œ ì„¤ì •
regions = ['ì„œìš¸', 'ë¶€ì‚°', 'ëŒ€êµ¬', 'ì¸ì²œ', 'ê´‘ì£¼', 'ëŒ€ì „', 'ìš¸ì‚°', 'ê°•ë¦‰']
dates = pd.date_range('2024-03-01', periods=5, freq='D')

# ëœë¤ ì‹œë“œ ì„¤ì •
np.random.seed(42)

# ë¹ˆ ë°ì´í„°í”„ë ˆì„ ìƒì„± í›„ ë°ì´í„° ì¶”ê°€
weather_df = pd.DataFrame()

# ê° ë‚ ì§œë³„ë¡œ ëª¨ë“  ì§€ì—­ ë°ì´í„° ìƒì„±
data_list = []
for date in dates:
    for region in regions:
        # ì˜¨ë„ (ì§€ì—­ë³„ íŠ¹ì„± ë°˜ì˜)
        base_temp = {'ì„œìš¸': 20, 'ë¶€ì‚°': 22, 'ëŒ€êµ¬': 21, 'ì¸ì²œ': 19,
                    'ê´‘ì£¼': 21, 'ëŒ€ì „': 20, 'ìš¸ì‚°': 23, 'ê°•ë¦‰': 18}
        temperature = base_temp[region] + np.random.normal(0, 3)

        # ê°•ìˆ˜ëŸ‰
        rainfall = max(0, np.random.normal(2, 8))
        if np.random.random() > 0.7:  # 30% í™•ë¥ ë¡œ ë¹„
            rainfall = max(rainfall, np.random.uniform(5, 30))
        else:
            rainfall = 0

        # ë‚ ì”¨ ìƒíƒœ
        if rainfall > 10:
            weather_status = 'ë¹„'
        elif rainfall > 0:
            weather_status = 'íë¦¼'
        else:
            weather_status = np.random.choice(['ë§‘ìŒ', 'íë¦¼'], p=[0.8, 0.2])

        data_list.append({
            'ë‚ ì§œ': date,
            'ì§€ì—­': region,
            'ì˜¨ë„': round(temperature, 1),
            'ê°•ìˆ˜ëŸ‰': round(rainfall, 1),
            'ë‚ ì”¨': weather_status
        })

# ë°ì´í„°í”„ë ˆì„ ìƒì„±
weather_df = pd.DataFrame(data_list)

print(" ë°ì´í„°í”„ë ˆì„ ìƒì„± ì™„ë£Œ!")
print(f" ë°ì´í„°í”„ë ˆì„ í¬ê¸°: {weather_df.shape} (í–‰: {weather_df.shape[0]}, ì—´: {weather_df.shape[1]})")
print()

# ë°ì´í„°í”„ë ˆì„ ê¸°ë³¸ ì •ë³´ í™•ì¸
print(" ë°ì´í„°í”„ë ˆì„ ê¸°ë³¸ ì •ë³´:")
print(weather_df.info())
print()

print(" ë°ì´í„° íƒ€ì… í™•ì¸:")
print(weather_df.dtypes)
print()

print(" ì²˜ìŒ 5í–‰ í™•ì¸:")
print(weather_df.head())
print()

print(" ë§ˆì§€ë§‰ 3í–‰ í™•ì¸:")
print(weather_df.tail(3))
print()
```

```python
# íŠ¹ì • ì»¬ëŸ¼ ì„ íƒ
print("ğŸ” íŠ¹ì • ì»¬ëŸ¼ ì„ íƒ:")
print("ì˜¨ë„ ì»¬ëŸ¼ë§Œ:")
print(weather_df['ì˜¨ë„'].head())
print()

print("ì—¬ëŸ¬ ì»¬ëŸ¼ ì„ íƒ:")
print(weather_df[['ì§€ì—­', 'ì˜¨ë„', 'ë‚ ì”¨']].head())
print()

# í–‰ ì„ íƒ (.loc ì‚¬ìš©)
print(" .locë¥¼ ì´ìš©í•œ í–‰ ì„ íƒ:")
print("ì„œìš¸ ë°ì´í„°ë§Œ:")
seoul_data = weather_df.loc[weather_df['ì§€ì—­'] == 'ì„œìš¸']
print(seoul_data)
print()

# ì¡°ê±´ë¶€ ì„ íƒ
print(" ì¡°ê±´ë¶€ ë°ì´í„° ì„ íƒ:")
hot_days = weather_df.loc[weather_df['ì˜¨ë„'] >= 23]
print(f"23ë„ ì´ìƒì¸ ë‚ : {len(hot_days)}ê°œ")
print(hot_days[['ë‚ ì§œ', 'ì§€ì—­', 'ì˜¨ë„', 'ë‚ ì”¨']].head())
print()

# ë³µí•© ì¡°ê±´
print(" ë³µí•© ì¡°ê±´ ì„ íƒ:")
rainy_hot = weather_df.loc[(weather_df['ì˜¨ë„'] >= 20) & (weather_df['ë‚ ì”¨'] == 'ë¹„')]
print(f"20ë„ ì´ìƒì´ë©´ì„œ ë¹„ ì˜¨ ê²½ìš°: {len(rainy_hot)}ê°œ")
print(rainy_hot[['ë‚ ì§œ', 'ì§€ì—­', 'ì˜¨ë„', 'ê°•ìˆ˜ëŸ‰']])
print()
```

```python
# ê²°ì¸¡ì¹˜ ì¶”ê°€ (ë°ì´í„°í”„ë ˆì„ ë³µì‚¬)
df_missing = weather_df.copy()

# ë¬´ì‘ìœ„ë¡œ ê²°ì¸¡ì¹˜ ì¶”ê°€
np.random.seed(123)
missing_indices = np.random.choice(len(df_missing), 8, replace=False)

# ì˜¨ë„ ê²°ì¸¡ì¹˜
df_missing.loc[missing_indices[:4], 'ì˜¨ë„'] = np.nan
# ê°•ìˆ˜ëŸ‰ ê²°ì¸¡ì¹˜
df_missing.loc[missing_indices[4:], 'ê°•ìˆ˜ëŸ‰'] = np.nan

print(" ê²°ì¸¡ì¹˜ í˜„í™©:")
print(df_missing.isnull().sum())
print()

print(" ê²°ì¸¡ì¹˜ê°€ ìˆëŠ” í–‰ë“¤:")
missing_rows = df_missing.loc[df_missing.isnull().any(axis=1)]
print(missing_rows[['ë‚ ì§œ', 'ì§€ì—­', 'ì˜¨ë„', 'ê°•ìˆ˜ëŸ‰', 'ë‚ ì”¨']])
print()

# ê²°ì¸¡ì¹˜ ì²˜ë¦¬ ë°©ë²•ë“¤
print(" ê²°ì¸¡ì¹˜ ì²˜ë¦¬ ë°©ë²•ë“¤:")

# ë°©ë²• 1: í‰ê· ê°’ìœ¼ë¡œ ì±„ìš°ê¸°
df_method1 = df_missing.copy()
temp_mean = df_method1['ì˜¨ë„'].mean()
rain_mean = df_method1['ê°•ìˆ˜ëŸ‰'].mean()
df_method1['ì˜¨ë„'].fillna(temp_mean, inplace=True)
df_method1['ê°•ìˆ˜ëŸ‰'].fillna(rain_mean, inplace=True)
print(f"ë°©ë²•1) í‰ê· ê°’ ì±„ìš°ê¸° - ì˜¨ë„: {temp_mean:.1f}ë„, ê°•ìˆ˜ëŸ‰: {rain_mean:.1f}mm")

# ë°©ë²• 2: ì§€ì—­ë³„ í‰ê· ìœ¼ë¡œ ì±„ìš°ê¸°
df_method2 = df_missing.copy()
df_method2['ì˜¨ë„'] = df_method2.groupby('ì§€ì—­')['ì˜¨ë„'].transform(lambda x: x.fillna(x.mean()))
df_method2['ê°•ìˆ˜ëŸ‰'] = df_method2.groupby('ì§€ì—­')['ê°•ìˆ˜ëŸ‰'].transform(lambda x: x.fillna(x.mean()))
print("ë°©ë²•2) ì§€ì—­ë³„ í‰ê· ìœ¼ë¡œ ì±„ìš°ê¸°")



# ìµœì¢… ë°ì´í„°ëŠ” ë°©ë²•2 ì‚¬ìš©
final_df = df_method2.copy()
```

```python
# ì§€ì—­ë³„ ì§‘ê³„
print(" ì§€ì—­ë³„ ê°•ìˆ˜ëŸ‰ í†µê³„2:")
region_stats = final_df.groupby('ì§€ì—­').agg({
    'ê°•ìˆ˜ëŸ‰': ['sum', 'mean'],
}).round(1)

# ì»¬ëŸ¼ëª… ì •ë¦¬
# region_stats.columns = ['ì´ê°•ìˆ˜ëŸ‰', 'í‰ê· ê°•ìˆ˜ëŸ‰']
print(region_stats)
print()
```

```python
# ì§€ì—­ë³„ ì§‘ê³„
print(" ì§€ì—­ë³„ í†µê³„:")
region_stats = final_df.groupby('ì§€ì—­').agg({
    'ì˜¨ë„': ['mean', 'max', 'min'],
    'ê°•ìˆ˜ëŸ‰': ['sum', 'mean'],
    'ë‚ ì”¨': 'count'
}).round(1)

# ì»¬ëŸ¼ëª… ì •ë¦¬
region_stats.columns = ['í‰ê· ì˜¨ë„', 'ìµœê³ ì˜¨ë„', 'ìµœì €ì˜¨ë„', 'ì´ê°•ìˆ˜ëŸ‰', 'í‰ê· ê°•ìˆ˜ëŸ‰', 'ë°ì´í„°ìˆ˜']
print(region_stats)
print()

# ë‚ ì”¨ë³„ ì§‘ê³„
print(" ë‚ ì”¨ë³„ í†µê³„:")
weather_stats = final_df.groupby('ë‚ ì”¨').agg({
    'ì˜¨ë„': 'mean',
    'ê°•ìˆ˜ëŸ‰': 'mean',
    'ì§€ì—­': 'count'
}).round(1)
weather_stats.columns = ['í‰ê· ì˜¨ë„', 'í‰ê· ê°•ìˆ˜ëŸ‰', 'ë°œìƒíšŸìˆ˜']
print(weather_stats)
print()

# ë‚ ì§œë³„ ì§‘ê³„
print("ë‚ ì§œë³„ ì „êµ­ í‰ê· :")
daily_stats = final_df.groupby('ë‚ ì§œ').agg({
    'ì˜¨ë„': 'mean',
    'ê°•ìˆ˜ëŸ‰': 'sum'
}).round(1)
daily_stats.columns = ['ì „êµ­í‰ê· ì˜¨ë„', 'ì „êµ­ì´ê°•ìˆ˜ëŸ‰']
print(daily_stats)
print()
```
