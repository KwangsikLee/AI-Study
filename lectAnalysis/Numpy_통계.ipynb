```python
import numpy as np
x = [18, 5, 10, 23, 19, -8, 10, 0, 0, 5, 2, 15, 8, 2, 5, 4, 15, -1, 4, -7, -24, 7, 9, -6, 23, -13]

print(len(x))                 #   : 26
print(sum(x))                 #  : 125
print(np.size(x))             #  : 26
print(np.sum(x))              #  : 125
print(np.mean(x))             #  : 4.8076923076923075
print(np.var(x))              #  : 115.23224852071006
print(np.std(x))              #  : 10.734628476137871
print(np.max(x))              #  : 23
print(np.min(x))              #  : -24
print(np.median(x))           #  : 5.0
print(np.percentile(x, 25))   # 1 : 0.0
print(np.percentile(x, 50))   # 2 : 5.0
print(np.percentile(x, 75))   # 3 : 10.0
print(np.percentile(x, 100))  # 4 : 23.0
```

# trim_mean

 TRIMMEAN(array, percent)             ,     .

  trimmean  , SciPy scipy.stats.trim_mean       

```python
from scipy import stats
x = [18, 5, 10, 23, 19, -8, 10, 0, 0, 5, 2, 15, 8, 2, 5, 4, 15, -1, 4, -7, -24, 7, 9, -6, 23, -13]

print(np.mean(x))                 #  : 4.8076923076923075
print(stats.trim_mean(x, 0.1))    #  : 5.2727272727272725   10% 
#print(stats.gmean(x))            #  :   0  
#print(stats.hmean(x))            #  :   0  
print(stats.skew(x))              #  : -0.4762339485461929
print(stats.kurtosis(x))          #  : 0.37443381660038977
print(stats.sem(x))               #  : 2.146925695227574
print(stats.describe(x))          #    (n, min, max, mean, var, skew, kurtosis)
dap = stats.describe(x)           #    
print(dap.mean)                   #   : 4.8076923076923075
```
