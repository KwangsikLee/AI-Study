```python
import numpy as np
x = [18, 5, 10, 23, 19, -8, 10, 0, 0, 5, 2, 15, 8, 2, 5, 4, 15, -1, 4, -7, -24, 7, 9, -6, 23, -13]

print(len(x))                 # 데이터 개수 : 26
print(sum(x))                 # 합계 : 125
print(np.size(x))             # 개수 : 26
print(np.sum(x))              # 합계 : 125
print(np.mean(x))             # 평균 : 4.8076923076923075
print(np.var(x))              # 분산 : 115.23224852071006
print(np.std(x))              # 표준편차 : 10.734628476137871
print(np.max(x))              # 최대값 : 23
print(np.min(x))              # 최소값 : -24
print(np.median(x))           # 중앙값 : 5.0
print(np.percentile(x, 25))   # 1사분위수 : 0.0
print(np.percentile(x, 50))   # 2사분위수 : 5.0
print(np.percentile(x, 75))   # 3사분위수 : 10.0
print(np.percentile(x, 100))  # 4사분위수 : 23.0
```

# trim_mean

엑셀의 TRIMMEAN(array, percent) 함수는 전체 데이터에서 가장 큰 값과 가장 작은 값을 일정 비율만큼 잘라낸 다음, 남은 값들의 평균을 구하는 함수이다.

넘파이 단독으로는 trimmean이라는 함수는 없지만, SciPy의 scipy.stats.trim_mean 함수를 사용하면 동일한 기능을 사용할 수 있다

```python
from scipy import stats
x = [18, 5, 10, 23, 19, -8, 10, 0, 0, 5, 2, 15, 8, 2, 5, 4, 15, -1, 4, -7, -24, 7, 9, -6, 23, -13]

print(np.mean(x))                 # 평균 : 4.8076923076923075
print(stats.trim_mean(x, 0.1))    # 절삭평균 : 5.2727272727272725 → 양쪽 10%씩 제거
#print(stats.gmean(x))            # 기하평균 : 음수 또는 0이 포함되면 에러
#print(stats.hmean(x))            # 조화평균 : 음수 또는 0이 포함되면 에러
print(stats.skew(x))              # 왜도 : -0.4762339485461929
print(stats.kurtosis(x))          # 첨도 : 0.37443381660038977
print(stats.sem(x))               # 표준오차 : 2.146925695227574
print(stats.describe(x))          # 전체 통계 요약 (n, min, max, mean, var, skew, kurtosis)
dap = stats.describe(x)           # 요약 통계 객체 생성
print(dap.mean)                   # 평균만 출력 : 4.8076923076923075
```
