```python
import torch, numpy as np, random
from transformers import AutoTokenizer, AutoModelForCausalLM

SEED = 123
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

MODEL_NAME = "skt/kogpt2-base-v2"
tok = AutoTokenizer.from_pretrained(MODEL_NAME)
mdl = AutoModelForCausalLM.from_pretrained(MODEL_NAME).to(DEVICE)

def generate_text(prompt, **gen_kwargs):
    ids = tok(prompt, return_tensors="pt").to(DEVICE)
    out = mdl.generate(**ids, **gen_kwargs)
    return tok.decode(out[0], skip_special_tokens=True)

prompt = "딥러닝 언어 모델의 핵심 아이디어는"
common = dict(max_new_tokens=80, repetition_penalty=1.1, eos_token_id=tok.eos_token_id)

print("=== Greedy ===")
print(generate_text(prompt, **common, do_sample=False))

print("\n=== Top-k(50) ===")
print(generate_text(prompt, **common, do_sample=True, top_k=50, temperature=0.9))

print("\n=== Nucleus(top_p=0.92) ===")
print(generate_text(prompt, **common, do_sample=True, top_p=0.92, temperature=0.8))
```
