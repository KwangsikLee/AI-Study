```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
import matplotlib.pyplot as plt
import re
from collections import Counter

# 한글 폰트 설정 (Colab용)
plt.rcParams['font.family'] = 'DejaVu Sans'

print("TensorFlow 버전:", tf.__version__)

# 한글 학습 데이터 - 한국 속담과 명언
korean_proverbs = """
가는 말이 고와야 오는 말이 곱다.
티끌 모아 태산이다.
천리 길도 한 걸음부터다.
누워서 떡 먹기라고 했지만 실제로는 어렵다.
백문이 불여일견이라고 하니 직접 해보자.
호랑이도 제 말 하면 온다.
열 번 찍어 안 넘어가는 나무 없다.
원숭이도 나무에서 떨어진다.
고생 끝에 낙이 온다.
소 잃고 외양간 고치기보다 미리 준비하자.
꿈을 포기하지 말고 끝까지 노력하자.
작은 일부터 차근차근 해나가자.
오늘 할 일을 내일로 미루지 말자.
친구와 함께하면 어려운 일도 쉬워진다.
정직이 최선의 정책이다.
시간은 금이라고 했으니 소중히 쓰자.
건강이 최고의 재산이다.
배움에는 끝이 없다고 하니 계속 공부하자.
실패는 성공의 어머니라고 했다.
웃음은 만병통치약이라고 한다.
진정한 친구는 어려울 때 알 수 있다.
노력하면 반드시 결실을 맺는다.
인내는 쓰지만 그 열매는 달다.
겸손한 마음을 항상 가지자.
감사하는 마음을 잊지 말자.
"""

class KoreanRNNGenerator:
    def __init__(self, text):
        self.original_text = text
        self.text = self.preprocess_text(text)

        # 문자 단위로 처리
        self.chars = sorted(list(set(self.text)))
        self.char_to_idx = {char: idx for idx, char in enumerate(self.chars)}
        self.idx_to_char = {idx: char for idx, char in enumerate(self.chars)}
        self.vocab_size = len(self.chars)

        print(f"총 텍스트 길이: {len(self.text):,}")
        print(f"고유 문자 수: {self.vocab_size}")
        print(f"문자 예시: {self.chars[:20]}")

    def preprocess_text(self, text):
        # 기본 전처리
        text = text.strip()
        text = re.sub(r'\s+', ' ', text)  # 여러 공백을 하나로
        return text.lower()

    def create_sequences(self, seq_length=30):
        sequences = []
        next_chars = []

        # 슬라이딩 윈도우로 시퀀스 생성
        for i in range(len(self.text) - seq_length):
            sequences.append(self.text[i:i + seq_length])
            next_chars.append(self.text[i + seq_length])

        print(f"총 시퀀스 수: {len(sequences):,}")

        # 원-핫 인코딩
        X = np.zeros((len(sequences), seq_length, self.vocab_size), dtype=np.bool_)
        y = np.zeros((len(sequences), self.vocab_size), dtype=np.bool_)

        for i, seq in enumerate(sequences):
            for t, char in enumerate(seq):
                X[i, t, self.char_to_idx[char]] = 1
            y[i, self.char_to_idx[next_chars[i]]] = 1

        return X, y

    def build_model(self, seq_length=30, rnn_units=128):
        model = Sequential([
            SimpleRNN(rnn_units,
                     input_shape=(seq_length, self.vocab_size),
                     dropout=0.2,
                     recurrent_dropout=0.2,
                     return_sequences=False),
            Dense(64, activation='relu'),
            Dropout(0.3),
            Dense(self.vocab_size, activation='softmax')
        ])

        model.compile(
            optimizer=Adam(learning_rate=0.002),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        return model

    def sample_next_char(self, predictions, temperature=1.0):
        """온도를 적용한 확률적 샘플링"""
        predictions = np.asarray(predictions).astype('float64')
        predictions = np.log(predictions + 1e-8) / temperature
        exp_preds = np.exp(predictions)
        predictions = exp_preds / np.sum(exp_preds)
        probas = np.random.multinomial(1, predictions, 1)
        return np.argmax(probas)

    def generate_text(self, model, seed_text, length=200, temperature=0.8, seq_length=30):
        """텍스트 생성"""
        generated = seed_text.lower()

        for _ in range(length):
            # 시드에서 마지막 seq_length만큼 사용
            if len(generated) >= seq_length:
                sequence = generated[-seq_length:]
            else:
                sequence = generated.ljust(seq_length)

            # 시퀀스를 원-핫으로 변환
            x_pred = np.zeros((1, seq_length, self.vocab_size))
            for t, char in enumerate(sequence):
                if char in self.char_to_idx:
                    x_pred[0, t, self.char_to_idx[char]] = 1

            # 예측
            predictions = model.predict(x_pred, verbose=0)[0]
            next_idx = self.sample_next_char(predictions, temperature)
            next_char = self.idx_to_char[next_idx]

            generated += next_char

        return generated

def run_rnn_korean_example():
    print("=" * 50)
    print("예제 1: RNN으로 한글 문장 생성")
    print("=" * 50)

    # 1. 데이터 준비
    generator = KoreanRNNGenerator(korean_proverbs)

    seq_length = 30
    X, y = generator.create_sequences(seq_length)

    print(f"\n데이터 형태:")
    print(f"X shape: {X.shape}")
    print(f"y shape: {y.shape}")

    # 2. 모델 구축
    model = generator.build_model(seq_length, rnn_units=128)
    model.summary()

    # 3. 콜백 설정
    early_stopping = EarlyStopping(
        monitor='loss',
        patience=5,
        restore_best_weights=True
    )

    # 4. 모델 훈련
    print("\n모델 훈련 시작...")
    history = model.fit(
        X, y,
        epochs=100,
        batch_size=128,
        callbacks=[early_stopping],
        verbose=1
    )

    # 5. 훈련 과정 시각화
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # 6. 텍스트 생성 테스트
    print("\n" + "="*50)
    print("생성된 한글 텍스트")
    print("="*50)

    seed_texts = ["가는 말이", "꿈을 포기", "친구와 함께"]
    temperatures = [0.5, 0.8, 1.2]

    for seed in seed_texts:
        print(f"\n🌱 시드 텍스트: '{seed}'")
        print("-" * 40)

        for temp in temperatures:
            generated = generator.generate_text(
                model, seed, length=100, temperature=temp, seq_length=seq_length
            )
            print(f"온도 {temp}: {generated[:150]}...")
            print()

    return model, generator, history

# 실행
if __name__ == "__main__":
    model, generator, history = run_rnn_korean_example()
```
