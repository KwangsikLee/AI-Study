```python
# 1. 설치
!pip install transformers evaluate rouge_score bert-score -q

# 2. 라이브러리 불러오기
from transformers import pipeline
import evaluate

# 3. 샘플 문서 (길게 작성된 텍스트)
document = """
Artificial intelligence (AI) is intelligence demonstrated by machines,
as opposed to the natural intelligence displayed by humans and animals.
Leading AI textbooks define the field as the study of intelligent agents:
any device that perceives its environment and takes actions that maximize
its chance of achieving its goals.
"""

# 참조 요약 (Reference)
reference_summary = "AI is intelligence shown by machines, studied as intelligent agents."

# 4. 모델 불러오기 (요약 전용 모델 - BART)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# 5. 모델 요약 생성
pred_summary = summarizer(document, max_length=40, min_length=10, do_sample=False)[0]["summary_text"]
print("📌 모델 요약:", pred_summary)
print("📌 참조 요약:", reference_summary)

# 6. 평가 지표 불러오기
rouge = evaluate.load("rouge")
bleu = evaluate.load("bleu")
bertscore = evaluate.load("bertscore")

# 7. ROUGE Score
rouge_score = rouge.compute(predictions=[pred_summary], references=[reference_summary])
print("\nROUGE Score:", rouge_score)

# 8. BLEU Score
bleu_score = bleu.compute(predictions=[pred_summary], references=[[reference_summary]])
print("\nBLEU Score:", bleu_score)

# 9. BERTScore
bertscore_result = bertscore.compute(predictions=[pred_summary], references=[reference_summary], lang="en")
print("\nBERTScore (F1):", bertscore_result["f1"][0])
```
