```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=10)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

font_path='/System/Library/Fonts/AppleGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=10)
plt.rcParams['font.family'] = 'AppleGothic'  # macOS
plt.rcParams['axes.unicode_minus'] = False
data_path = './data/geo/'
```

```python
# 1. CIFAR-10 샘플 이미지
(x_train, _), _ = tf.keras.datasets.cifar10.load_data()
img = x_train[0]
img = tf.convert_to_tensor(img)

# 2. 흑백(Grayscale) 변환
gray = tf.image.rgb_to_grayscale(img)

# 3. Grayscale → RGB 변환 (채널 복제)
rgb = tf.image.grayscale_to_rgb(gray)

# 4. 컬러맵(Jet 등)으로 컬러처럼 시각화
plt.figure(figsize=(12,3))
plt.subplot(1,4,1)
plt.imshow(img.numpy())
plt.title('원본 RGB')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(gray.numpy().squeeze(), cmap='gray')
plt.title('Grayscale')
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(rgb.numpy())
plt.title('복원된 RGB\n(사실상 흑백)')
plt.axis('off')

# 컬러맵을 사용해 가짜 컬러 이미지처럼 표현
plt.subplot(1,4,4)
# plt.imshow(gray.numpy().squeeze(), cmap='jet')
plt.imshow(gray.numpy(), cmap='jet')
plt.title('컬러맵(Jet)')
plt.axis('off')

plt.suptitle("흑백 → RGB 변환 및 컬러맵 시각화")
plt.show()
```
