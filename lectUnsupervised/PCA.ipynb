```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=10)
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
digits = load_digits()
X = digits.data
y = digits.target
# print(digits.feature_names)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='tab10', s=15, alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Digits 데이터 PCA(2D) 시각화")
plt.colorbar(label='Digit label')
plt.show()

print(f"PC1 설명 분산: {pca.explained_variance_ratio_[0]:.3f}")
print(f"PC2 설명 분산: {pca.explained_variance_ratio_[1]:.3f}")
print(f"합계: {pca.explained_variance_ratio_[:2].sum():.3f}")
```

```python
# 1. 데이터 및 라이브러리 준비
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd

# 2. 데이터 로드
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# 3. 2차원 PCA 적용
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 4. 결과 DataFrame으로 보기 좋게
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df_pca['target'] = y

# 5. 시각화
plt.figure(figsize=(7,5))
for i, name in enumerate(target_names):
    plt.scatter(df_pca.loc[df_pca['target']==i, 'PC1'],
                df_pca.loc[df_pca['target']==i, 'PC2'],
                label=name, alpha=0.7)
plt.xlabel("PC1 (주성분1)")
plt.ylabel("PC2 (주성분2)")
plt.title("Iris 데이터 PCA(2D) 시각화")
plt.legend()
plt.show()

# 6. 설명 분산 비율
print(f"주성분 1이 설명하는 분산: {pca.explained_variance_ratio_[0]:.3f}")
print(f"주성분 2가 설명하는 분산: {pca.explained_variance_ratio_[1]:.3f}")
print(f"두 주성분 합계: {pca.explained_variance_ratio_[:2].sum():.3f}")
```
