```python
# World Bank API에서 국가 정보를 가져오는 예제
import requests
import json

# 1. API 엔드포인트 설정
url = "https://api.worldbank.org/v2/countries?format=json"

# 2. GET 요청 보내기
response = requests.get(url)

# 3. 상태 코드 확인
if response.status_code == 200:
    print(" 데이터 요청 성공!")
else:
    print(" 데이터 요청 실패! 상태코드:", response.status_code)

# 4. JSON 데이터 파싱
data = response.json()

# 5. 전체 데이터 구조 확인 (리스트 형태로 반환됨)
print("\n전체 데이터 구조 예시:")
print(json.dumps(data[0], indent=2))  # 페이지 정보
print(json.dumps(data[1][0], indent=2))  # 첫 번째 국가 정보 예시

# 6. 국가 이름과 ISO 코드만 추출하여 출력
print("\n 국가 리스트 (이름 - ISO 코드):")
for country in data[1]:
    print(f"{country['name']} - {country['id']}")
```
