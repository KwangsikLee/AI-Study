```python
%%writefile app.py
# 파일명 예: app.py

import streamlit as st
import pandas as pd
import folium
from folium.plugins import HeatMap
from streamlit_folium import folium_static

st.title("이디야 매장 히트맵 + 마커")

df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 지도 중심 설정
center = [35.15805, 129.0823]
m = folium.Map(location=center, zoom_start=12)

# 히트맵
heat_data = df[['위도', '경도']].values.tolist()
HeatMap(heat_data, radius=15, blur=10).add_to(m)

# 마커 추가
for i in df.index:
    folium.Marker(
        location=df.loc[i, ['위도', '경도']],
        tooltip=df.loc[i, '상호명'],
        icon=folium.Icon(color='blue', icon='coffee', prefix='fa')
    ).add_to(m)

folium_static(m)
```

```python
%%writefile app1.py
import streamlit as st
import pandas as pd
import folium
from folium.plugins import HeatMap
from streamlit_folium import folium_static
from folium.plugins import HeatMap, MarkerCluster

# CSV 데이터 로드
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 지도 초기화
center = [35.15805, 129.0823]
map = folium.Map(location=center, zoom_start=12)

# 히트맵 추가
heat_data = df[['위도', '경도']].values.tolist()
HeatMap(heat_data, radius=15).add_to(map)

# MarkerCluster 생성
marker_cluster = MarkerCluster().add_to(map)

# 마커 반복 추가
for i in df.index:
    name = df.loc[i, '상호명']
    lat = df.loc[i, '위도']
    lon = df.loc[i, '경도']

    # 팝업 HTML 구성
    popup_html = f"<b>{name}</b><br>위도: {lat}<br>경도: {lon}"

    folium.Marker(
        location=[lat, lon],
        popup=folium.Popup(popup_html, max_width=250),
        tooltip=name,
        icon=folium.Icon(color='blue', icon='coffee', prefix='fa')
    ).add_to(marker_cluster)

# 저장
#map.save('ediya_map_with_cluster.html')
folium_static(map)
```

```python
%%writefile app2.py
import streamlit as st
import pandas as pd
import folium
from folium.plugins import HeatMap
from streamlit_folium import folium_static
from folium.plugins import HeatMap, MarkerCluster

# 1. CSV 파일 불러오기
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 2. 주소에서 '구' 정보 추출
gu_list = []
for address in df['주소']:
    parts = address.split()
    if len(parts) > 1:
        gu_list.append(parts[1])
    else:
        gu_list.append('')
df['구'] = gu_list

# 3. 사용자 입력 받기 (검색어)
search_gu = st.text_input("검색할 구 이름을 입력하세요 (예: 수영): ","")

#  4. 필터링 (검색어가 포함된 구 이름 검색, 대소문자 무시)
filtered_df = df[df['구'].str.contains(search_gu, case=False)].reset_index(drop=True)

#  5. 검색 결과가 없는 경우 처리
if filtered_df.empty:
    st.warning(f"'{search_gu}'을(를) 포함한 동 이름을 가진 매장이 없습니다.")
else:
    st.warning(f"{len(filtered_df)}개의 매장이 검색되었습니다.")

    #  6. 지도 중심: 첫 번째 매장 위치
    center = [filtered_df.loc[0, '위도'], filtered_df.loc[0, '경도']]
    map_filtered = folium.Map(location=center, zoom_start=13)

    #  7. 마커 추가
    for i in filtered_df.index:
        name = filtered_df.loc[i, '상호명']
        lat = filtered_df.loc[i, '위도']
        lon = filtered_df.loc[i, '경도']

        folium.Marker(
            location=[lat, lon],
            tooltip=name,
            popup=folium.Popup(name, max_width=200),
            icon=folium.Icon(color='blue', icon='coffee', prefix='fa')
        ).add_to(map_filtered)

    #  8. 지도 저장
    #output_path = f"/content/ediya_{search_gu}_map.html"
    #map_filtered.save(output_path)
    #print(f"지도가 HTML로 저장되었습니다: {output_path}")

    folium_static(map_filtered)
```
