```python
import pandas as pd
import folium

# 1. CSV 파일 불러오기
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 2. 주소에서 '구' 정보 추출
gu_list = []
for address in df['주소']:
    parts = address.split()
    if len(parts) > 1:
        gu_list.append(parts[1])
    else:
        gu_list.append('')
df['구'] = gu_list

# 3. 사용자 입력 받기 (검색어)
search_gu = input("검색할 구 이름을 입력하세요 (예: 수영): ").strip()

#  4. 필터링 (검색어가 포함된 구 이름 검색, 대소문자 무시)
filtered_df = df[df['구'].str.contains(search_gu, case=False)].reset_index(drop=True)

#  5. 검색 결과가 없는 경우 처리
if filtered_df.empty:
    print(f"'{search_gu}'을(를) 포함한 구 이름을 가진 매장이 없습니다.")
else:
    print(f"{len(filtered_df)}개의 매장이 검색되었습니다.")

    #  6. 지도 중심: 첫 번째 매장 위치
    center = [filtered_df.loc[0, '위도'], filtered_df.loc[0, '경도']]
    map_filtered = folium.Map(location=center, zoom_start=13)

    #  7. 마커 추가
    for i in filtered_df.index:
        name = filtered_df.loc[i, '상호명']
        lat = filtered_df.loc[i, '위도']
        lon = filtered_df.loc[i, '경도']

        folium.Marker(
            location=[lat, lon],
            tooltip=name,
            popup=folium.Popup(name, max_width=200),
            icon=folium.Icon(color='blue', icon='coffee', prefix='fa')
        ).add_to(map_filtered)

    #  8. 지도 저장
    output_path = f"/content/ediya_{search_gu}_map.html"
    map_filtered.save(output_path)
    print(f"지도가 HTML로 저장되었습니다: {output_path}")

    map_filtered
```
