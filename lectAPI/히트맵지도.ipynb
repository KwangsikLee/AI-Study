```python
import folium
import pandas as pd
from folium.plugins import HeatMap

# CSV 파일 로드
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 지도 중심 좌표 설정
center = [35.15805, 129.0823]
map = folium.Map(location=center, zoom_start=12)

# 히트맵용 데이터 생성: 위도, 경도 쌍으로 리스트 구성
heat_data = df[['위도', '경도']].values.tolist()


# 히트맵 추가
HeatMap(heat_data, radius=15, blur=10, min_opacity=0.3, max_zoom=15,
        gradient = {
            0.2: 'blue',
            0.4: 'lime',
            0.6: 'orange',
            0.8: 'red'
        }).add_to(map)

map
```

```python
import folium
import pandas as pd
from folium.plugins import HeatMap

# CSV 파일 로드
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 지도 중심 좌표 설정
center = [35.15805, 129.0823]
map = folium.Map(location=center, zoom_start=12)

# 히트맵 데이터 생성
heat_data = df[['위도', '경도']].values.tolist()
HeatMap(heat_data, radius=15, blur=10).add_to(map)

# 마커 추가 (커피 아이콘)
for i in df.index:
    if df.loc[i, '상호명'] == '이디야':
      popup_html = f"<br>상호:{df.loc[i, '상호명']}"
      folium.Marker(
          location=df.loc[i, ['위도', '경도']],
          tooltip=df.loc[i, '상호명'],
          popup=folium.Popup(popup_html, max_width=300),
          icon=folium.Icon(color='blue', icon='coffee', prefix='fa')
      ).add_to(map)

# map.save('map.html')
map
```
