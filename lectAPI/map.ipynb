```python
# 지도에 표시할 기준점을 잡아준다.
center = [35.15805, 129.0823]
map = folium.Map(location=center, zoom_start=12)

# 데이터 파일에서 불러온 100개의 데이터에서 위도, 경로에 대한 값을 지도에 표시한다.
for i in df.index[:100]:
    folium.Circle(
        location = df.loc[i, ['latitude', 'longitude']],
        tooltip = df.loc[i, 'name'],
        radius = 100,
        color = 'red',
        fill_color = 'red'
      ).add_to(map)
map
```

```python
import folium
import pandas as pd

# 데이터 파일에서 불러온다.
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')
df.head()

# 지도에 표시할 기준점을 잡아준다.
center = [35.15805, 129.0823]
map = folium.Map(location=center, zoom_start=12)

# 데이터 파일에 저장된 위도, 경로에 대한 값을 지도에 표시한다.
for i in df.index[:len(df)]:      # 데이터 개수만큼 반복한다.
    folium.Circle(
        location = df.loc[i, ['위도', '경도']],
        tooltip = df.loc[i, '상호명'],
        radius = 100,
        color = 'blue',
        fill_color = 'blue'
      ).add_to(map)
map
```

```python
import folium
import pandas as pd

# CSV 파일 로드 (한글 인코딩 주의)
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 지도 중심 좌표 설정
center = [35.15805, 129.0823]
map = folium.Map(location=center, zoom_start=12)

# 마커로 표시
for i in df.index:
    folium.Marker(
        location=df.loc[i, ['위도', '경도']],
        tooltip=df.loc[i, '상호명'],  # 마우스 오버 시 표시
        icon=folium.Icon(color='blue', icon='info-sign')  # 아이콘 스타일 설정
    ).add_to(map)

# 지도 출력
map
```

```python
import folium
import pandas as pd

# CSV 파일 로드
df = pd.read_csv('/content/EDIYA.csv', encoding='CP949')

# 지도 중심 좌표 설정
center = [35.15805, 129.0823]
map = folium.Map(location=center, zoom_start=12)

# 마커를 커피 아이콘으로 표시
for i in df.index:
    folium.Marker(
        location=df.loc[i, ['위도', '경도']],
        tooltip=df.loc[i, '상호명'],
        icon=folium.Icon(color='blue', icon='coffee', prefix='fa')  # Font Awesome 아이콘 사용
    ).add_to(map)

map
```
