```python

# ============================================================================
# ÏòàÏ†ú 2: Ï§ëÍ∏â CNNÏùÑ Ïù¥Ïö©Ìïú CIFAR-10 Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄ Î∂ÑÎ•ò
# ============================================================================

# CIFAR-10 ÌÅ¥ÎûòÏä§ Ïù¥Î¶Ñ
CLASS_NAMES = ['ÎπÑÌñâÍ∏∞', 'ÏûêÎèôÏ∞®', 'ÏÉà', 'Í≥†ÏñëÏù¥', 'ÏÇ¨Ïä¥',
               'Í∞ú', 'Í∞úÍµ¨Î¶¨', 'Îßê', 'Î∞∞', 'Ìä∏Îü≠']

def load_and_preprocess_cifar10():
    """CIFAR-10 Îç∞Ïù¥ÌÑ∞Î•º Î°úÎìúÌïòÍ≥† Ï†ÑÏ≤òÎ¶¨Ìï©ÎãàÎã§."""
    print(" CIFAR-10 Îç∞Ïù¥ÌÑ∞ Î°úÎî© Ï§ë...")

    # CIFAR-10 Îç∞Ïù¥ÌÑ∞ÏÖã Î°úÎìú
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

    print(f"ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÌòïÌÉú: {x_train.shape}")
    print(f"ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÌòïÌÉú: {x_test.shape}")
    print(f"Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞: {x_train.shape[1:3]}, Ï±ÑÎÑê: {x_train.shape[3]}")

    # Ï†ïÍ∑úÌôî (0-255 -> 0-1)
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0

    # Ïõê-Ìï´ Ïù∏ÏΩîÎî©
    y_train = tf.keras.utils.to_categorical(y_train, 10)
    y_test = tf.keras.utils.to_categorical(y_test, 10)

    return (x_train, y_train), (x_test, y_test)

def visualize_cifar10_samples(x_data, y_data, num_samples=10):
    """CIFAR-10 ÏÉòÌîå Ïù¥ÎØ∏ÏßÄÎì§ÏùÑ ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§."""
    plt.figure(figsize=(15, 3))
    for i in range(num_samples):
        plt.subplot(2, 5, i + 1)
        plt.imshow(x_data[i])
        plt.title(f'{CLASS_NAMES[np.argmax(y_data[i])]}')
        plt.axis('off')
    plt.suptitle('CIFAR-10 ÏÉòÌîå Ïù¥ÎØ∏ÏßÄÎì§', fontsize=16)
    plt.tight_layout()
    plt.show()

def create_improved_cnn():
    """Í∞úÏÑ†Îêú CNN Î™®Îç∏ÏùÑ ÏÉùÏÑ±Ìï©ÎãàÎã§."""
    print(" Í∞úÏÑ†Îêú CNN Î™®Îç∏ Íµ¨ÏÑ± Ï§ë...")

    model = models.Sequential([
        # Ï≤´ Î≤àÏß∏ Conv Block
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Îëê Î≤àÏß∏ Conv Block
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # ÏÑ∏ Î≤àÏß∏ Conv Block
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.25),

        # Fully Connected Layers
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])

    return model

def create_data_augmentation():
    """Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ïÏùÑ ÏúÑÌïú ImageDataGeneratorÎ•º ÏÉùÏÑ±Ìï©ÎãàÎã§."""
    print(" Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï ÏÑ§Ï†ï Ï§ë...")

    datagen = ImageDataGenerator(
        rotation_range=15,          # 15ÎèÑ ÌöåÏ†Ñ
        width_shift_range=0.1,      # Í∞ÄÎ°ú Ïù¥Îèô
        height_shift_range=0.1,     # ÏÑ∏Î°ú Ïù¥Îèô
        horizontal_flip=True,       # Ï¢åÏö∞ Î∞òÏ†Ñ
        zoom_range=0.1,             # ÌôïÎåÄ/Ï∂ïÏÜå
        shear_range=0.1,            # Ï†ÑÎã® Î≥ÄÌôò
        fill_mode='nearest'         # Îπà Í≥µÍ∞Ñ Ï±ÑÏö∞Í∏∞
    )

    return datagen

def visualize_data_augmentation(datagen, x_train):
    """Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï ÏòàÏãúÎ•º ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§."""
    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ ÏÑ†ÌÉù
    sample_img = x_train[0:1]  # Ï≤´ Î≤àÏß∏ Ïù¥ÎØ∏ÏßÄ

    plt.figure(figsize=(15, 3))

    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄ
    plt.subplot(1, 6, 1)
    plt.imshow(sample_img[0])
    plt.title('ÏõêÎ≥∏')
    plt.axis('off')

    # Ï¶ùÍ∞ïÎêú Ïù¥ÎØ∏ÏßÄÎì§
    aug_iter = datagen.flow(sample_img, batch_size=1)
    for i in range(5):
        plt.subplot(1, 6, i + 2)
        aug_img = next(aug_iter)[0]
        plt.imshow(aug_img)
        plt.title(f'Ï¶ùÍ∞ï {i+1}')
        plt.axis('off')

    plt.suptitle('Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï ÏòàÏãú', fontsize=16)
    plt.tight_layout()
    plt.show()

def train_with_callbacks(model, train_data, validation_data, datagen, epochs=20):
    """ÏΩúÎ∞±ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Î™®Îç∏ÏùÑ ÌõàÎ†®Ìï©ÎãàÎã§."""
    print(f"\n Î™®Îç∏ ÌõàÎ†® ÏãúÏûë (ÏóêÌè¨ÌÅ¨: {epochs})...")

    # Î™®Îç∏ Ïª¥ÌååÏùº
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # ÏΩúÎ∞± ÏÑ§Ï†ï
    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=5,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=0.0001
        )
    ]

    # Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ïÏùÑ ÏÇ¨Ïö©Ìïú ÌõàÎ†®
    history = model.fit(
        datagen.flow(train_data[0], train_data[1], batch_size=32),
        steps_per_epoch=len(train_data[0]) // 32,
        epochs=epochs,
        validation_data=validation_data,
        callbacks=callbacks,
        verbose=1
    )

    return history

def create_confusion_matrix(model, test_data):
    """ÌòºÎèô ÌñâÎ†¨ÏùÑ ÏÉùÏÑ±ÌïòÍ≥† ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§."""
    from sklearn.metrics import confusion_matrix, classification_report
    import seaborn as sns

    # ÏòàÏ∏° ÏàòÌñâ
    y_pred = model.predict(test_data[0])
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(test_data[1], axis=1)

    # ÌòºÎèô ÌñâÎ†¨ ÏÉùÏÑ±
    cm = confusion_matrix(y_true_classes, y_pred_classes)

    # ÌòºÎèô ÌñâÎ†¨ ÏãúÍ∞ÅÌôî
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
    plt.title('ÌòºÎèô ÌñâÎ†¨ (Confusion Matrix)')
    plt.xlabel('ÏòàÏ∏° Î†àÏù¥Î∏î')
    plt.ylabel('Ïã§Ï†ú Î†àÏù¥Î∏î')
    plt.xticks(rotation=45)
    plt.yticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Î∂ÑÎ•ò Î¶¨Ìè¨Ìä∏ Ï∂úÎ†•
    print("\n Î∂ÑÎ•ò Î¶¨Ìè¨Ìä∏:")
    report = classification_report(y_true_classes, y_pred_classes,
                                 target_names=CLASS_NAMES)
    print(report)

def analyze_predictions(model, test_data, num_samples=10):
    """ÏòàÏ∏° Í≤∞Í≥ºÎ•º Î∂ÑÏÑùÌïòÍ≥† ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§."""
    predictions = model.predict(test_data[0][:num_samples])

    fig, axes = plt.subplots(2, 5, figsize=(15, 8))
    for i in range(num_samples):
        ax = axes[i//5, i%5]
        ax.imshow(test_data[0][i])

        # ÏòàÏ∏° Í≤∞Í≥º
        pred_probs = predictions[i]
        true_label = np.argmax(test_data[1][i])
        pred_label = np.argmax(pred_probs)
        confidence = np.max(pred_probs)

        # ÏÉâÏÉÅ ÏÑ§Ï†ï (Ï†ïÎãµ: Ï¥àÎ°ù, Ïò§Îãµ: Îπ®Í∞ï)
        color = 'green' if true_label == pred_label else 'red'

        ax.set_title(f'Ïã§Ï†ú: {CLASS_NAMES[true_label]}\n'
                    f'ÏòàÏ∏°: {CLASS_NAMES[pred_label]}\n'
                    f'Ïã†Î¢∞ÎèÑ: {confidence:.2f}',
                    color=color, fontsize=10)
        ax.axis('off')

    plt.suptitle('CIFAR-10 ÏòàÏ∏° Í≤∞Í≥º Î∂ÑÏÑù', fontsize=16)
    plt.tight_layout()
    plt.show()

# ============================================================================
# Î©îÏù∏ Ïã§Ìñâ Ìï®Ïàò
# ============================================================================

def main():
    """CNN ÏòàÏ†ú 2Ïùò Î©îÏù∏ Ïã§Ìñâ Ìï®ÏàòÏûÖÎãàÎã§."""
    print(" CNN ÏòàÏ†ú 2: CIFAR-10 Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄ Î∂ÑÎ•ò")
    print("=" * 50)

    # 1. Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï†ÑÏ≤òÎ¶¨
    (x_train, y_train), (x_test, y_test) = load_and_preprocess_cifar10()

    # 2. ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî
    visualize_cifar10_samples(x_train, y_train)

    # 3. Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï ÏÑ§Ï†ï
    datagen = create_data_augmentation()
    datagen.fit(x_train)

    # 4. Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï ÏòàÏãú ÏãúÍ∞ÅÌôî
    visualize_data_augmentation(datagen, x_train)

    # 5. Î™®Îç∏ ÏÉùÏÑ±
    model = create_improved_cnn()
    model.summary()

    # 6. Î™®Îç∏ ÌõàÎ†®
    history = train_with_callbacks(model, (x_train, y_train),
                                 (x_test, y_test), datagen, epochs=20)

    # 7. ÌõàÎ†® Í≥ºÏ†ï ÏãúÍ∞ÅÌôî
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    # Ï†ïÌôïÎèÑ
    ax1.plot(history.history['accuracy'], label='accuracy')
    ax1.plot(history.history['val_accuracy'], label='val_accuracy')
    ax1.set_title('model accuracy')
    ax1.set_xlabel('epoch')
    ax1.set_ylabel('accuracy')
    ax1.legend()
    ax1.grid(True)

    # ÏÜêÏã§
    ax2.plot(history.history['loss'], label='loss')
    ax2.plot(history.history['val_loss'], label='val_loss')
    ax2.set_title('model loss')
    ax2.set_xlabel('epoch')
    ax2.set_ylabel('epoch')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

    # 8. Î™®Îç∏ ÌèâÍ∞Ä
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    print(f"\n test_accuracy: {test_accuracy:.4f}")

    # 9. ÌòºÎèô ÌñâÎ†¨ ÏÉùÏÑ±
    create_confusion_matrix(model, (x_test, y_test))

    # 10. ÏòàÏ∏° Í≤∞Í≥º Î∂ÑÏÑù
    analyze_predictions(model, (x_test, y_test))

    return model

# Ïã§Ìñâ
if __name__ == "__main__":
    trained_model = main()
```
