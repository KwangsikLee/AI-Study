```python

# ================== 데이터 ==================
import pandas as pd

data = [
    {
        "title":"인셉션",
        "genres":"SF 스릴러 액션",
        "overview":"타인의 꿈 속에 들어가 아이디어를 심는 팀의 위험한 작전",
        "actors":["레오나르도 디카프리오","조셉 고든 레빗","엘렌 페이지"],
        "director":"크리스토퍼 놀란",
        "keywords":["꿈","잠입","의식","하이스트"]
    },
    {
        "title":"그래비티",
        "genres":"SF 스릴러",
        "overview":"우주에서 고립된 두 우주비행사의 생존 사투",
        "actors":["산드라 블록","조지 클루니"],
        "director":"알폰소 쿠아론",
        "keywords":["우주","고립","생존","재난"]
    },
    {
        "title":"매드맥스 분노의 도로",
        "genres":"액션 모험",
        "overview":"사막의 종말 세계에서 자유를 향한 카체이스",
        "actors":["톰 하디","샤를리즈 테론","니콜라스 홀트"],
        "director":"조지 밀러",
        "keywords":["사막","추격전","해방","포스트아포칼립스"]
    },
    {
        "title":"라라랜드",
        "genres":"드라마 로맨스 음악",
        "overview":"재즈 피아니스트와 배우 지망생의 사랑과 꿈",
        "actors":["라이언 고슬링","엠마 스톤"],
        "director":"데이미언 셔젤",
        "keywords":["재즈","꿈","사랑","로스앤젤레스"]
    },
    {
        "title":"기생충",
        "genres":"드라마 스릴러",
        "overview":"가난한 가족이 부유한 집에 스며들며 벌어지는 블랙코미디",
        "actors":["송강호","이선균","조여정","최우식","박소담"],
        "director":"봉준호",
        "keywords":["계급","기만","비밀","반전"]
    },
    {
        "title":"옥자",
        "genres":"드라마 모험",
        "overview":"거대 동물 옥자를 지키기 위해 소녀가 떠나는 모험",
        "actors":["안서현","틸다 스윈튼","제이크 질렌할"],
        "director":"봉준호",
        "keywords":["우정","동물권","기업","환경"]
    },
    {
        "title":"부산행",
        "genres":"액션 스릴러 좀비",
        "overview":"좀비가 창궐한 열차에서 벌어지는 사투",
        "actors":["공유","정유미","마동석","김수안"],
        "director":"연상호",
        "keywords":["좀비","생존","열차","감염"]
    },
    {
        "title":"위플래쉬",
        "genres":"드라마 음악",
        "overview":"천재 드러머 지망생과 혹독한 스승의 집착과 대결",
        "actors":["마일즈 텔러","J.K. 시몬스"],
        "director":"데이미언 셔젤",
        "keywords":["음악","집착","스승과제자","재즈드럼"]
    },
    {
        "title":"셰이프 오브 워터",
        "genres":"드라마 판타지 로맨스",
        "overview":"말 못 하는 여성과 수중 생명체의 특별한 사랑",
        "actors":["샐리 호킨스","마이클 섀넌","더그 존스"],
        "director":"기예르모 델 토로",
        "keywords":["수중","비밀연구소","사랑","다름"]
    },
    {
        "title":"닥터 스트레인지",
        "genres":"SF 액션 판타지",
        "overview":"교만한 천재 외과의가 마법사가 되어 세상을 위협으로부터 지킨다",
        "actors":["베네딕트 컴버배치","틸다 스윈튼","레이첼 맥아담스"],
        "director":"스콧 데릭슨",
        "keywords":["마법","차원","시간조작","히어로"]
    },
    {
        "title":"아바타",
        "genres":"SF 모험 판타지",
        "overview":"판도라 행성에서 벌어지는 인간과 나비족의 갈등과 화해",
        "actors":["샘 워싱턴","조 샐다나","시고니 위버"],
        "director":"제임스 카메론",
        "keywords":["외계행성","환경","전쟁","사랑"]
    },
    {
        "title":"타이타닉",
        "genres":"드라마 로맨스 재난",
        "overview":"거대한 여객선 타이타닉호의 침몰과 두 사람의 사랑",
        "actors":["레오나르도 디카프리오","케이트 윈슬렛"],
        "director":"제임스 카메론",
        "keywords":["사랑","침몰","계급","희생"]
    },
    {
        "title":"블랙 팬서",
        "genres":"액션 SF 히어로",
        "overview":"와칸다의 왕이자 블랙 팬서가 되어 나라와 세계를 지킨다",
        "actors":["채드윅 보스만","루피타 뇽오","마이클 B. 조던"],
        "director":"라이언 쿠글러",
        "keywords":["히어로","왕국","기술","전통"]
    },
    {
        "title":"겨울왕국",
        "genres":"애니메이션 가족 판타지",
        "overview":"얼음을 다루는 힘을 가진 언니와 동생의 모험과 사랑",
        "actors":["크리스틴 벨","이디나 멘젤","조시 게드"],
        "director":"크리스 벅, 제니퍼 리",
        "keywords":["자매","얼음","왕국","자유"]
    },
    {
        "title":"해리포터와 마법사의 돌",
        "genres":"판타지 모험",
        "overview":"평범한 소년이 마법학교에 입학하며 벌어지는 이야기",
        "actors":["다니엘 래드클리프","엠마 왓슨","루퍼트 그린트"],
        "director":"크리스 콜럼버스",
        "keywords":["마법","학교","우정","운명"]
    },
    {
        "title":"반지의 제왕: 반지 원정대",
        "genres":"판타지 모험",
        "overview":"절대반지를 파괴하기 위해 모인 원정대의 여정",
        "actors":["일라이저 우드","이안 맥켈런","비고 모텐슨"],
        "director":"피터 잭슨",
        "keywords":["반지","모험","우정","희생"]
    },
    {
        "title":"조커",
        "genres":"드라마 스릴러 범죄",
        "overview":"한 남자가 광대에서 고담의 악당 조커로 변모하는 이야기",
        "actors":["호아킨 피닉스","로버트 드 니로","재지 비츠"],
        "director":"토드 필립스",
        "keywords":["고독","광기","사회문제","변화"]
    },
    {
        "title":"어벤져스: 엔드게임",
        "genres":"액션 SF 히어로",
        "overview":"타노스에 맞서 인류를 구하기 위해 히어로들이 다시 뭉친다",
        "actors":["로버트 다우니 주니어","크리스 에반스","스칼렛 요한슨"],
        "director":"안소니 루소, 조 루소",
        "keywords":["히어로","희생","시간여행","전투"]
    },
    {
        "title":"쇼생크 탈출",
        "genres":"드라마",
        "overview":"억울하게 수감된 남자가 감옥에서 희망을 잃지 않고 탈출을 꿈꾼다",
        "actors":["팀 로빈스","모건 프리먼"],
        "director":"프랭크 다라본트",
        "keywords":["감옥","희망","우정","탈출"]
    },
    {
        "title":"캐러비안의 해적: 블랙펄의 저주",
        "genres":"모험 액션 판타지",
        "overview":"해적 잭 스패로우가 블랙펄호와 전설의 저주를 되찾으려 한다",
        "actors":["조니 뎁","올란도 블룸","키이라 나이틀리"],
        "director":"고어 버빈스키",
        "keywords":["해적","저주","모험","보물"]
    }
]

df = pd.DataFrame(data)

# ================== 형태소 분석기 (Mecab 우선, 실패 시 Okt) ==================
from konlpy.tag import Okt
tokenizer_name = "okt"
okt = Okt()
mecab = None
try:
    from konlpy.tag import Mecab
    mecab = Mecab()  # 코랩은 mecab-ko 사전이 기본 설치 안 돼도 mecab-python3로 동작
    tokenizer_name = "mecab"
except Exception as e:
    print("[알림] Mecab 사용 불가 → Okt로 진행합니다.")
print("사용 토크나이저:", tokenizer_name)

# ================== 불용어 ==================
stopwords = set("""
은 는 이 가 을 를 에 에서 으로 도 와 과 의 로 으로부터 부터 보다 그리고 그러나 하지만 또한
하다 되다 이다 있다 없다 모든 같은 매우 너무 가장 대한 위해 으며 으면서 등 각 해당 또 또는
""".split())
# 도메인 상투어/장르명 약간 추가 (원하면 빼세요)
stopwords.update("""
드라마 스릴러 액션 SF 모험 코미디 로맨스 판타지 애니메이션 가족 음악 영화 작품 이야기 장면
""".split())

# ================== 토크나이저 ==================
def tokenize_ko(text: str):
    if not isinstance(text, str): return []
    text = text.strip()
    if tokenizer_name == "mecab":
        tokens = mecab.morphs(text)
    else:
        tokens = okt.morphs(text, stem=True)
    out = []
    for t in tokens:
        t = t.lower()
        if t in stopwords:
            continue
        if len(t) < 2:
            continue
        out.append(t)
    return out

# ================== 멀티소스 텍스트 + 가중치 ==================
# 장르/줄거리/배우/감독/키워드 가중치 조절
W_GENRES   = 2.0
W_OVERVIEW = 1.0
W_ACTORS   = 1.5
W_DIRECTOR = 1.2
W_KEYWORDS = 1.8

def build_weighted_text(row):
    # 리스트 필드는 공백으로 합침
    actors   = " ".join(row.get("actors", []) or [])
    keywords = " ".join(row.get("keywords", []) or [])
    genres   = row.get("genres", "") or ""
    overview = row.get("overview", "") or ""
    director = row.get("director", "") or ""
    # 단순 반복 대신, 가중치만큼 토큰을 복제해서 넣는 방식(간단하고 효과적)
    # ex) W=2.0이면 2번 반복, 1.5면 1번 반복 + 나머지는 확률적으로 더해도 되지만 여기선 반올림
    def repeat(text, w):
        n = max(1, int(round(w)))
        return (" " + text) * n if text else ""
    
    return (repeat(genres, W_GENRES) +
            repeat(overview, W_OVERVIEW) +
            repeat(actors, W_ACTORS) +
            repeat(director, W_DIRECTOR) +
            repeat(keywords, W_KEYWORDS)).strip()

df["raw_text"] = df.apply(build_weighted_text, axis=1)

# 형태소 전처리 → 토큰 문자열
df["tokens"] = df["raw_text"].map(tokenize_ko)
df["text_for_tfidf"] = df["tokens"].map(lambda toks: " ".join(toks))

df
```

```python

# ================== TF-IDF & 유사도 ==================
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(token_pattern=r"[^ ]+", ngram_range=(1,2), max_features=10000)
tfidf = vectorizer.fit_transform(df["text_for_tfidf"])
cosine_sim = cosine_similarity(tfidf, tfidf)

# ================== 추천 함수 ==================
from difflib import get_close_matches

title_to_idx = {t.lower(): i for i, t in enumerate(df["title"])}

def _resolve_title(user_title: str):
    key = user_title.strip().lower()
    if key in title_to_idx:
        return key
    cand = get_close_matches(key, list(title_to_idx.keys()), n=1, cutoff=0.5)
    return cand[0] if cand else None

def recommend(title: str, top_n: int = 5, include_score: bool = True):
    key = _resolve_title(title)
    if not key:
        print(f"[경고] '{title}' 제목을 찾지 못했습니다.")
        return pd.DataFrame()
    idx = title_to_idx[key]
    sims = list(enumerate(cosine_sim[idx]))
    sims = sorted([(i, s) for i, s in sims if i != idx], key=lambda x: -x[1])[:top_n]
    rec_idx = [i for i, _ in sims]
    out = df.loc[rec_idx, ["title","genres","overview","actors","director","keywords"]].copy()
    if include_score:
        out["similarity"] = [round(s, 4) for _, s in sims]
    return out.reset_index(drop=True)

# ================== 사용 예시 ==================
print("=== [인셉션] 비슷한 영화 ===")
display(recommend("인셉션", top_n=3))

print("=== [라라랜드] 비슷한 영화 ===")
display(recommend("라라랜드", top_n=3))

print("=== [기생충] 비슷한 영화 ===")
display(recommend("기생충", top_n=3))
```
